### Handle null values in betting columns

```python
betting_columns = [
    "live_bets_count",
    "avg_bet_amount",
    "total_turnover",
    "total_payout",
    "ggr",
    "ggr_margin"
]

df_churn[betting_columns] = df_churn[betting_columns].fillna(0)
```

### Handling missing values in deposit columns

```python
df_churn["deposit_count"] = df_churn["deposit_count"].fillna(0)
df_churn["total_deposit_amount"] = df_churn["total_deposit_amount"].fillna(0)
```

```python
df_churn["deposit_to_turnover_ratio"] = df_churn["total_deposit_amount"] / df_churn["total_turnover"].replace(0, np.nan)
df_churn["deposit_to_turnover_ratio"] = df_churn["deposit_to_turnover_ratio"].fillna(0)
```

### Inspect the rest of missing data

```python
df_missing_live_dates = df_churn[
    df_churn["first_live_bet_date"].isna() &
    df_churn["last_live_bet_date"].isna()
]
```

## Data types conversion

```python
# Convert date columns from string to date in order to generate features from them

date_columns = [
    "registration_date",
    "observation_start_date",
    "observation_end_date",
    "last_live_bet_date",
    "first_live_bet_date"
]

for col in date_columns:
    df_churn[col] = pd.to_datetime(df_churn[col], errors="coerce")
```

## Feature engineering

```python
df_churn["days_since_last_bet"] = (df_churn["observation_end_date"] - df_churn["last_live_bet_date"]).dt.days
df_churn["days_since_last_bet"] = df_churn["days_since_last_bet"].fillna(999)
```

```python
df_churn["days_since_last_bet"]
```

```python
df_churn["observation_days"] = (
    df_churn["observation_end_date"] - df_churn["observation_start_date"]
).dt.days

df_churn["daily_bet_activity_rate"] = df_churn["days_active_in_observation"] / df_churn["observation_days"]
```

```python
df_churn["daily_bet_activity_rate"]
```

```python
df_churn["inactivity_rate"] = df_churn["days_since_last_bet"] / df_churn["tenure_days"]
```

```python
df_churn["ggr_per_bet"] = df_churn["ggr"] / df_churn["live_bets_count"].replace(0, np.nan)
df_churn["ggr_per_bet"] = df_churn["ggr_per_bet"].fillna(0)
```

```python
df_churn["deposit_per_bet"] = df_churn["total_deposit_amount"] / df_churn["live_bets_count"].replace(0, np.nan)
df_churn["deposit_per_bet"] = df_churn["deposit_per_bet"].fillna(0)
```

```python
df_churn["days_to_first_bet"] = (df_churn["first_live_bet_date"] - df_churn["registration_date"]).dt.days
df_churn["days_to_first_bet"] = df_churn["days_to_first_bet"].fillna(-1)
```

```python
df_churn["betting_window_days"] = (df_churn["last_live_bet_date"] - df_churn["first_live_bet_date"]).dt.days
df_churn["betting_window_days"] = df_churn["betting_window_days"].fillna(0)
```

```python
df_churn["activity_intensity"] = df_churn["live_bets_count"] / df_churn["tenure_days"].replace(0, np.nan)
df_churn["activity_intensity"] = df_churn["activity_intensity"].fillna(0)
```

```python
df_churn["compact_turnover"] = np.log1p(df_churn["total_turnover"]) 
df_churn["compact_deposit"] = np.log1p(df_churn["total_deposit_amount"])
```

```python
df_churn.to_csv("updated_churn_dataset.csv", index=False)
```
